```python
import os
import json
import re
from tqdm import tqdm
```

```python
config_dir = "/home/zjusst/qms/webarena/config_files"

reddit_ids = []

file_names = os.listdir(config_dir)

for file_name in tqdm(file_names):
    if file_name.endswith(".json"):
        if "test" in file_name:
            continue
        if int(file_name[:-5]) <= 811:
            with open(os.path.join(config_dir, file_name), "r") as f:
                config = json.load(f)
                sites = config["sites"]
                task_id = config["task_id"]
                if sites == ["reddit"]:
                    reddit_ids.append(task_id)
        else:
            #print("skipping", file_name)
            pass
    else:
        #print("skipping", file_name)    
        pass
    #print(config)
```

```python
def find_continuous_ranges(nums):
    # 先排序，去重
    nums = sorted(set(nums))
    if not nums:
        return []
    ranges = []
    start = prev = nums[0]
    for n in nums[1:]:
        if n == prev + 1:
            prev = n
            continue
        else:
            ranges.append([start, prev])
            start = prev = n
    ranges.append([start, prev])
    return ranges

reddit_ranges = find_continuous_ranges(reddit_ids)
print(reddit_ranges)
```
