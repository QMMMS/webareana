```python
import json

osgensis_datas = []
with open("/home/zjusst/qms/osgensis/os_genesis_web_training.jsonl", "r") as f:
    for line in f:
        data = json.loads(line)
        osgensis_datas.append(data)

print(len(osgensis_datas))
osgensis_datas[0]
```

```python
from pprint import pprint
import copy

last_instruction = ""
trajs = []
temp_traj = []

for index in range(len(osgensis_datas)):
    user_input = osgensis_datas[index]['conversations'][0]['value']
    instruction = ""
    for line in user_input.split('\n'):
        if line.startswith('Instruction:'):
            instruction = line

    if instruction != last_instruction:
        if last_instruction != "":
            trajs.append(copy.deepcopy(temp_traj))
            temp_traj = []
        last_instruction = instruction

    temp_traj.append(osgensis_datas[index])

print(len(trajs))
```

```python
for index in range(len(trajs)):
    # str_index 4 index str 
    str_index = str(index).zfill(4)
    
    out_file_path = f"/home/zjusst/qms/osgensis/total_trajs/traj_{str_index}.json"

    with open(out_file_path, "w", encoding="utf-8") as f:
        json.dump(trajs[index], f, ensure_ascii=False, indent=4)

    print(f"Saved {out_file_path}")

print("All trajectories have been saved.")
```
