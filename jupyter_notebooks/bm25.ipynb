```python
# 你的文档库（字符串组成的列表）
corpus = [
    "A RAG system combines a retriever with a large language model.",
    "BM25 is a powerful keyword-based retrieval algorithm.",
    "Vector search finds semantically similar documents using embeddings.",
    "Hybrid search often combines the strengths of BM25 and vector search.",
    "The retriever's job is to find relevant documents for the language model."
]

# 你的查询字符串
query = "retriever for language model"
```

```python
from rank_bm25 import BM25Okapi

tokenized_corpus = [doc.split(" ") for doc in corpus]
bm25 = BM25Okapi(tokenized_corpus)
tokenized_query = query.split(" ")

top_n_docs = bm25.get_top_n(tokenized_query, corpus, n=3)
top_n_docs_index = [corpus.index(doc) for doc in top_n_docs]

print(top_n_docs_index)

for doc in top_n_docs:
    print(doc)
```

```python
doc_scores = bm25.get_scores(tokenized_query)
print("\nScores for each document:")
print(doc_scores)
```
