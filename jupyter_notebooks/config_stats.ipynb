```python
# 列出 /home/zjusst/qms/webarena/config_files 目录下的所有文件
import os
import json



reddit_only_case = []
wiki_only_case = []
reddit_other_case = []
wiki_other_case = []


files = os.listdir('/home/zjusst/qms/webarena/config_files')
for file in files:
    if file.endswith('.json') and "test" not in file:
        with open(os.path.join('/home/zjusst/qms/webarena/config_files', file), 'r') as f:
            data = json.load(f)

            try:
                if 'reddit' in data["sites"]:
                    if len(data["sites"]) == 1:
                        reddit_only_case.append(data)
                    else:
                        reddit_other_case.append(data)

                if 'wikipedia' in data["sites"]:
                    if len(data["sites"]) == 1:
                        wiki_only_case.append(data)
                    else:
                        wiki_other_case.append(data)
            except Exception as e:
                print(f"Error processing file: {file}")
                raise e

            # print(f"File: {file}, Keys: {list(data.keys())}")
```

```python
len(reddit_only_case), len(wiki_only_case), len(reddit_other_case), len(wiki_other_case)
```

```python
template_list = []
start_urls = []
task_ids = []
for case in reddit_only_case:
    if case["intent_template"] not in template_list:
        template_list.append(case["intent_template"])
    if case["start_url"] not in start_urls:
        start_urls.append(case["start_url"])
    if case["task_id"] not in task_ids:
        task_ids.append(case["task_id"])
len(template_list), len(start_urls), len(task_ids)
```

```python
task_ids.sort()
task_ids


# 将 task_ids int列表划分为不同的 列表，每个列表的元素都是连续的
def split_into_consecutive_groups(task_ids):
    if not task_ids:
        return []

    task_ids.sort()  # 确保列表是有序的
    groups = [[task_ids[0]]]  # 初始化第一个分组

    for i in range(1, len(task_ids)):
        if task_ids[i] == task_ids[i - 1] + 1:
            # 如果当前元素与前一个元素连续，加入当前分组
            groups[-1].append(task_ids[i])
        else:
            # 否则，开启一个新的分组
            groups.append([task_ids[i]])

    return groups

groups = split_into_consecutive_groups(task_ids)
groups
```
